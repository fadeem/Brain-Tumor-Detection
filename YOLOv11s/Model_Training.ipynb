```python
from ultralytics import YOLO
```

```python
new_model_path = "models"
new_model_name = "brain_tumor_yolo_v11s"

model_path = 'models/yolo11s.pt' #existing model to be fine tuned
model = YOLO(model_path)
```

```python
# Fine-tuning with your dataset
data_yaml = r"./dataset/data.yaml"
model.train(
    data=data_yaml,
    epochs=150, #100 epochs for good result
    imgsz=640,
    batch=32,
    workers=0,  # Use the main process for data loading
    name=new_model_name,
    project=new_model_path,
)

# model.train(
#     data=data_yaml,
#     epochs=200,                # Roboflow usually trains long enough
#     imgsz=640,                # Roboflow often uses 1024x1024 for aerial imagery
#     batch=32,                  # 32 might be too large unless you have a powerful GPU
#     workers=2,                 # Use 2+ workers unless Windows causes issues
#     name=new_model_name,
#     project=new_model_path,
#     lr0=0.01,                  # Learning rate — default is usually fine
#     lrf=0.01,                  # Final LR fraction (for cosine LR decay)
#     cos_lr=True,               # Enable cosine learning rate scheduler (Roboflow often uses this)
#     augment=True,              # Enable full augmentations
#     mosaic=1.0,                # Enable mosaic (on by default)
#     hsv_h=0.015,
#     hsv_s=0.7,
#     hsv_v=0.4,
#     flipud=0.0,
#     fliplr=0.5,
#     mixup=0.2,
#     translate=0.1,
#     scale=0.5,
#     degrees=0.0,
#     perspective=0.0,
#     copy_paste=0.0,
#     pretrained=True,          # Start from pretrained weights (default behavior)
#     device='0',               # GPU device (set to 'cpu' if no GPU)
#     patience=50,              # Early stopping patience
#     verbose=True              # Print detailed training logs
# )
```
